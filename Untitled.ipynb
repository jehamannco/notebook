```python
import sys
sys.path.append('../functions/')
from databases import *
from scipy.signal import medfilt, find_peaks, argrelextrema
import numpy as np
def get_peaks(df):
    """
    Function to get index values where there are peaks in the data (maxima and minima)
    df: dataframe containing the data to a analize
    """
    # get maxima
    max_id =find_peaks(tesla[mask].hvSocActualDisplay.values)[0]
    #get minima
    min_id =find_peaks(-1*tesla[mask].hvSocActualDisplay.values)[0]
    
    peaks_id = np.append(min_id, max_id)
    #add initial and final points
    peaks_id = np.append(peaks_id,[0,len(df)-1])
    peaks_id.sort()
    
    return peaks_id
```

```python
#load tesla data
tesla = load_base2_ftp('ecu_bcddc2ceb734')
#tesla = load_base2_ftp('ecu_bcddc2d0718c')

tesla = tesla[tesla.model == 'TESLA_S']
#tesla = tesla[tesla.model == 'KIAENIRO']

#select while charging data, for Tesla the conditions are as below
mask = (tesla.hvBatteryCurrent > 0) & (tesla.rpmEmotor <=0) 
#mask = (tesla.hvBatteryCurrent < 0) & (tesla.speed <=0) 

#find peaks to extract charging curves
peaks_id = get_peaks(tesla[mask])

%matplotlib inline

# plot SOC and peaks_id to check if the peaks were correctly determined

plt.plot(tesla[mask].timestamp.iloc[peaks_id], tesla[mask].hvSocActualDisplay.iloc[peaks_id], 'o', c='red', alpha=0.5 )
plt.plot(tesla[mask].timestamp, tesla[mask].hvSocActualDisplay, '.--')
plt.tick_params(axis='x',labelrotation=30)
plt.ylabel('SoC (%)')
plt.tight_layout()
```
